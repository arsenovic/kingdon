# Torus Knots  with PGA Rotors
Using rotors to generate torus knots. 

```python
import numpy as np
from kingdon import Algebra
exp = lambda x: x.exp()
#from kingdon.numerical import exp as exp_num # work around
#exp = lambda x: exp_num(x, 30)
pga = Algebra(3,0,1,start_index=0) 
locals().update(pga.blades)


# parameters
thetas = np.linspace(0,1*np.pi,401)

r1,r2 = .2,.5
p ,q  = 2,3

Rs = [exp(-theta*e12*p)*exp(r2*e01)*exp(-theta*e13*q)*exp(r1*e01) for theta in thetas]

o      = e0.dual() # origin 
points = [k>>o for k in Rs]
pga.graph(
      *points ,
    grid=False, lineWidth=4,
)
```

```python
def f(theta, p, q, r1, r2):
    return exp(-theta*e12*p)*exp(r2*e01)*exp(-theta*e13*q)*exp(r1*e01)

a = lambda theta: f(theta, p=2, q=3, r1=0.2, r2=0.5)
from kingdon import calculus as calc
da = calc.d(pga, a)
das = [da(theta) for theta in thetas]  # evaluate derivative
```

```python

points = [k>>o for k in das]
pga.graph(
      *points ,
    grid=False, lineWidth=4,
)
```
